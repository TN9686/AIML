```python
#adding of the element
l = ['b','y','e']
l[1]='e'
l=l+['e']
print(l)
```

```python
#append of the element
l = ['b','y','e']
l[1]='e'
l=l+['e']
l.append('y')
print(l)
```

```python
l = ['b','y','e']
l[1]='e'
l=l+['e']
l.append('y')
l.pop('b')
print(l)
```

```python
#delete
l = ['b','y','e']
l[1]='e'
l=l+['e']
l.append('y')
l.pop(0)
print(l)
```
