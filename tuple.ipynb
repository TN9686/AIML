```python
my_tuple = (10,20,30,50,40)
print("tuple:",my_tuple)
```

```python
#index
my_tuple = (10,20,30,50,40)
index_of_20 = my_tuple.index(20)
print("Index of 20:", index_of_20)
```

```python
#count()
my_tuple = (10,20,30,50,40,50,60,78,)
count_10 = my_tuple.count(10)
print("count of 10:", count_10)
```

```python
#len()
my_tuple = (10, 20, 30, 40)
print(len(my_tuple))
```

```python
#max and min
my_tuple = (10, 20, 30, 50, 40, 50, 60, 78)
maximum_value = max(my_tuple)
minimum_value = min(my_tuple)
print("Maximum:", maximum_value)
print("Minimum:", minimum_value)
```

```python
#sum()
my_tuple = (10, 20, 30, 50, 40)
total = sum(my_tuple)
print(total)
```

```python
#Given a tuple of numbers and a target element, find and print all indexes where the target element occurs in the tuple.

numbers = (10, 20, 30, 20, 40, 20, 50)
target =20
indexes = []
for i in range(len(numbers)):
    if numbers[i] == target:
        indexes.append(i)
print("indexes:", indexes)
```

```python
#set 
my_set = {1,2,4,4,5}
print("initial set:", my_set)
print(type(my_set))
```

```python
#add
my_set = {1,2,4}
my_set.add(3)
print("after add:",my_set)
```

```python
#update
my_set = {1,2}
my_set.update((1,2,3),[5,6,7])
print("after update:",my_set)
```

```python
#remove
my_set = {1,2}
my_set.remove(2)
print("after remove:",my_set)
```

```python
#discard( doesnt throw an error)
my_set = {1,2}
my_set.discard(4)
print("after discard:",my_set)
```

```python
#pop(removes last item)

my_set = {1,2,6}
remove_item = my_set.pop()
print("removed item:",remove_item)
print("after pop:",my_set)
```

```python
#clear

my_set = {10, 20, 30, 40}
my_set.clear()
print("After clear():", my_set)
```

```python
#union
set1 = {1,2,3}
set2 = {3,4,5}
result = set1.union(set2)
print("union:",result)
```

```python
#intersection
set1 = {1,2,3}
set2 = {3,2,5}
result = set1.intersection(set2)
print("intersection:",result)
```

```python
#symmetric difference (set1)
set1 = {1,2,3}
set2 = {3,2,4}
result = set1.symmetric_difference(set2)
print("symmetric:", result)
```

```python
#difference
set1 = {1,2,3}
set2 = {3,2,4}
result = set1.difference(set2)
print("symmetric:", result)
```

```python
#isssubset
set1 = {1,2}
set2 = {1,2,4}
print("Is set1 subset of set2?",set1.issubset(set2))
```

```python
#issuperset
set1 = {1,2,4}
set2 = {1,2}
print("Is set1 superset of set2?",set1.issuperset(set2))
```

```python
#remove duplicates from a list using set
numbers = [1, 2, 3, 2, 4, 1, 5]
my_set = set()
my_set.update(numbers)
result = list(my_set)

print("after removing duplicates:", result)
```
